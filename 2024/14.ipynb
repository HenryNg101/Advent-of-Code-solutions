```python
import os
os.chdir('..')
from inputGetter import get_input
import re, math

content = get_input(2024, 14)[:-1]
```

```python
content = content.split('\n')
positions, velocities = [], []
max_x, max_y = 101, 103

for line in content:
    position, velocity = [loc.split('=')[1] for loc in line.split(' ')]
    positions.append([int(val) for val in position.split(',')])
    velocities.append([int(val) for val in velocity.split(',')])

'''
Check if an 2d array may contains a Christmas tree or not, by checking if 
there are multiple robots stay consecutively in the same row
'''
def possible_tree(grid_map, rows_counts):
    pattern = 'xxxxxxxxxx'
    reg_exp = re.compile(pattern)
    
    for id, row in enumerate(grid_map):
        if rows_counts[id] < len(pattern):
            continue
        str_val = ''.join(row)
        if reg_exp.search(str_val):
            return True
    return False

'''Do an iteration of travelling for all of robots'''
def travel(grid_map, rows_counts):
    for i in range(len(positions)):
        positions[i][0] = (positions[i][0] + velocities[i][0]) % max_x
        positions[i][1] = (positions[i][1] + velocities[i][1]) % max_y
        rows_counts[positions[i][1]] += 1
        grid_map[positions[i][1]][positions[i][0]] = 'x'
```

```python
iteration = 1
tree_found = False
while True:    
    grid_map = [['.' for _ in range(max_x)] for _ in range(max_y)]
    rows_counts = [0 for _ in range(max_y)]

    travel(grid_map, rows_counts)
    
    # Logic part A
    if iteration == 100:
        quadrons = [0 for _ in range(4)]
        for position in positions:
            x, y = position
            if x != max_x // 2 and y != max_y // 2:
                quadrons[(y < max_y // 2) * 2 + (x < max_x // 2)] += 1  # Calculate the quadron to count the robot into
                
        print(f'Part A: {math.prod(quadrons)}')
    
    # Logic part B
    tree_found |= possible_tree(grid_map, rows_counts)
    if tree_found:
        print(f'Part B: {iteration + 1}')
        for line in grid_map:
            print(''.join(line))
        
    if tree_found and iteration > 100:
        break
    
    iteration += 1
```
